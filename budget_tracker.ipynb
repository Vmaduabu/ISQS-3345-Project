```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("budget.csv")
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df = df.dropna(subset=["date"])

selected_month = "2025-10"

monthly_df = df[df["date"].dt.strftime("%Y-%m") == selected_month]

if not monthly_df.empty:
    total_income = monthly_df[monthly_df["type"].str.lower() == "income"]["amount"].sum()
    total_expense = monthly_df[monthly_df["type"].str.lower() == "expense"]["amount"].sum()
    balance = total_income - total_expense

    print("Monthly Budget Summary")
    print("---------------------------")
    print(f"Selected Month: {selected_month}")
    print(f"Total Income:   ${total_income:,.2f}")
    print(f"Total Expenses: ${total_expense:,.2f}")
    print(f"Net Balance:    ${balance:,.2f}")
else:
    print(f"No transactions found for {selected_month}. Check your date format or dataset.")
```

Step #4 Visualization charts: Now that onthly data has been filtered and summarized, this step visualizes the expenses by category.

Using Matplotlib and Seaborn, the chart below shows how much was spent in each category for the selected month.  
This helps identify where most of the money is going (For example: rent, groceries, utilities, transport, etc.).

```python
# Filter only expense data for that month
expense_data = monthly_df[monthly_df["type"].str.lower() == "expense"]

# Group total expenses by category
category_totals = expense_data.groupby("category")["amount"].sum().reset_index()

# Bar chart visualization
plt.figure(figsize=(8, 5))
sns.barplot(data=category_totals, x="category", y="amount", palette="viridis")

plt.title(f"Expenses by Category for {selected_month}")
plt.xlabel("Category")
plt.ylabel("Amount ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The Bar chart shows that rent is clearly the highest expense for that month, then followed bhy groceries, utilities, and transport. This visualization makes it easier to actually see the spending pattern
